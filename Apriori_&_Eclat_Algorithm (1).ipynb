```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
data = pd.read_csv("Market_Basket_Optimisation.csv", header = None)
```

```python
data
```

```python
# Create a for-loop to pass all the dataframe into an array
transactions = []
for i in range(0, 7501):
  transactions.append([str(data.values[i,j]) for j in range(0, 20)])
```

```python
# If we use the Eclat algorithm, we can keep the min_support and remove the rest but it is highly recommended to keep
# the rest of the parameters as they will contribute an even stronger association rules.
# Eclat is a set of frequent bought products, Apriori is more of Association rules.
from apyori import apriori
rules = apriori(transactions = transactions,
                min_support = 0.003,
                min_confidence = 0.2,
                min_lift = 3,
                min_length = 2,
                max_lenght = 2)
```

```python
results = list(rules)
results
```
